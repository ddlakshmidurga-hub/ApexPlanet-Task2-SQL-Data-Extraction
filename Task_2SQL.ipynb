```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("enhanced_zomato_dataset_clean.csv")
```

```python
df.head()
```

```python
conn = sqlite3.connect("zomato.db")
```

```python
df.to_sql(
    "zomato",
    conn,
    if_exists="replace",
    index=False
)
```

```python
pd.read_sql("SELECT * FROM zomato LIMIT 5;", conn)
```

```python
pd.read_sql("""
SELECT Restaurant_Name, City
FROM zomato;
""", conn)
```

```python
df.columns.tolist()
```

```python
pd.read_sql("""
SELECT *
FROM zomato
WHERE Average_Rating > 4.5;
""", conn)
```

```python
pd.read_sql("""
SELECT Restaurant_Name, Average_Rating
FROM zomato
ORDER BY Average_Rating DESC;
""", conn)
```

```python
pd.read_sql("""
SELECT Restaurant_Name, Average_Rating
FROM zomato
ORDER BY Average_Rating DESC
LIMIT 10;
""", conn)
```

```python
pd.read_sql("""
SELECT City, COUNT(*) AS Total_Restaurants
FROM zomato
GROUP BY City;
""", conn)
```

```python
pd.read_sql("""
SELECT City, AVG(Average_Rating) AS Avg_Rating
FROM zomato
GROUP BY City;
""", conn)
```
